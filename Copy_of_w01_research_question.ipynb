## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/rajavi-mhatre/flyrank-ml/main/data/raw/content_refresh_anonymized.csv")

# Number of pages with a downward direction trend.
print(df['trend_direction'].value_counts())

# Pages with downward direction trends AND high nuber of clicks.
decline = df[(df["trend_direction"]=="down") & (df['impressions_90d']>100)]
print(f"Declining-with-demand pages: {len(decline)} out of {len(df)} ({len(decline)/len(df)*100:.1f}%)")
```
